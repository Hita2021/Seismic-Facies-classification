# mode filter

```python
from scipy import stats
def patch_extractor2D(img,mid_x,mid_y,l,w):
    patch= np.pad(img,((l//2,l//2),(w//2,w//2)), 'edge')[mid_x:mid_x+l,mid_y:mid_y+w] 
    return patch

def find_mode(x):
    patch = np.reshape(x,(x.shape[0]*x.shape[1],1))
    mode = stats.mode(patch)[0][0][0]
    return mode

def mode_filter(img,l,w):
    output=np.zeros(np.shape(img),dtype=np.single)
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            patch_mode = find_mode(patch_extractor2D(img,x,y,l,w))
            output[x][y] = patch_mode
    return output   
```
